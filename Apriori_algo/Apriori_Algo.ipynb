```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('/content/OnlineRetail5000.csv' , encoding='iso-8859-1')
df
```

```python
df.dropna(axis = 0)
df = df.iloc[0:5000,:]
df
```

```python
# from google.colab import files

# df.to_csv('OnlineRetail5000.csv', index=False)
```

```python
df.head()
```

```python
df["Description"] = df['Description'].str.strip()
df.dropna(axis = 0 , subset=['InvoiceNo'], inplace = True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
df
```

```python
basket = (df[df['Country'] == 'United Kingdom']
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
          .unstack()
          .reset_index()
          .fillna(0)
          .set_index('InvoiceNo'))
basket
```

```python
def encode_units(x):
  if x<= 0:
    return 0
    if x >= 1:
      return 1
basket_sets = basket.applymap(encode_units)
# basket_sets.drop("POSTAGE", axis = 1, inplace  = True )
basket_sets

basket_sets.fillna(0, inplace=True)  # Fill NaN values with 0
basket_sets
```

```python
frequent_items = apriori(basket_sets, min_support=0.07, use_colnames=True)
print(frequent_items)
rules = association_rules(frequent_items, metric='lift', min_threshold=1)
```

```python
rules = [(rules['lift']>= 6) &(rules["confidence"] >= 0.8)]
```
