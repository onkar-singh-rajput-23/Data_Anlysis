```python
#Q 3
def Noise(Column_Name):
  if df[Column_Name].dtypes == np.float64:
    for i in df[Column_Name]:
      if type(i)==int or type(i)==str:
        df[Column_Name]=df[Column_Name].replace(i, 0)

  if df[Column_Name].dtypes == np.int64:
    for i in df[Column_Name]:
      if type(i)==int or type(i)==str:
        df[Column_Name]=df[Column_Name].replace(i, 0)

  if df[Column_Name].dtypes == "object":
    for i in df[Column_Name]:
      if type(i)==int or type(i)==str:
        df[Column_Name]=df[Column_Name].replace(i, 0)
Noise("salary")
Noise("job_title")
df
```
