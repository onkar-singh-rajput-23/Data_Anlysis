```python
#Some Important Libary
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

sns.set()

from sklearn import decomposition
from sklearn import datasets

from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#Q(1).
iris_dataset=pd.read_csv("/content/iris.data",names=["sepel length","sepel width","petal length","petal width","class"])
iris_dataset.head()
```

```python
iris_dataset.isnull()
```

```python
iris_dataset.loc[:,iris_dataset.isnull().any()].columns
```

```python
X = iris_dataset.iloc[:,:3]
```

```python
y= iris_dataset.iloc[:,4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=7)
```

```python
#Q(2)

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_pred, y_test)
print( accuracy*100)
```

```python
#Q(3)
pca = decomposition.PCA(n_components=3)
pca.fit(X_train)
```

```python
transformed_X = pca.transform(X_train)
```

```python
transformed_X
```

```python
#Q(4)
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(transformed_X, y_train)
```

```python
X_test_transformed = pca.transform(X_test)
```

```python
y_pred2 = knn.predict(X_test_transformed)
```

```python
accuracy = accuracy_score(y_pred2, y_test)
print(accuracy*100)
```

```python
#Q(5)
l =[]
for i in range(1,4):
  
  pca = decomposition.PCA(n_components=i)
  pca.fit(X_train)
  transformed_X = pca.transform(X_train)
  knn = KNeighborsClassifier(n_neighbors=4)
  knn.fit(transformed_X, y_train)
  X_test_transformed = pca.transform(X_test)
  y_pred2 = knn.predict(X_test_transformed)
  accuracy = accuracy_score(y_pred2, y_test)
  l.append(accuracy)
print(l)
```
