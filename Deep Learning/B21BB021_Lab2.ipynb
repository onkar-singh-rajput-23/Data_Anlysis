```python
class UnderCompleteAutoencoder(nn.Module):
    def __init__(self,input_dim):
        super(UnderCompleteAutoencoder, self).__init__()

        # Encoder layers
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, int(input_dim*.75)),
            nn.ReLU(),
            nn.Linear(int(input_dim*.75), int(input_dim*.5)),
            nn.ReLU(),
            nn.Linear(int(input_dim*.5), int(input_dim*.25)),
            nn.ReLU(),
        )

        # Decoder layers
        self.decoder = nn.Sequential(
            nn.Linear(int(input_dim*.25), int(input_dim*.5)),
            nn.ReLU(),
            nn.Linear(int(input_dim*.5), input_dim*.75),
            nn.ReLU(),
             nn.Linear(int(input_dim*.75), input_dim),
            nn.Sigmoid(),  # Sigmoid activation to squash output between 0 and 1
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
input_dim = 3*32*32
batch_size = 16
learning_rate= 0.001
epochs = 20
```

```python
import torch
#  use gpu if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# create a model from `AE` autoencoder class
# load it to the specified device, either gpu or cpu
model = UnderCompleteAutoencoder(input_dim).to(device)

# create an optimizer object
# Adam optimizer with learning rate 1e-3
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# mean-squared error loss
criterion = nn.MSELoss()
```
